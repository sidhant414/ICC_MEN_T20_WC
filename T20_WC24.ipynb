```python
matches_per_date = icc_t20_df['Date of Match'].value_counts()

matches_category_counts = matches_per_date.value_counts().reindex([1,2,3])

labels = ['1 Match', '2 Matches', '3 Matches']
sizes= matches_category_counts.values

plt.figure(figsize=(8,8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['skyblue', 'green', 'orange'])
plt.title('Matches per Day Category')
```

```python
# super over match column
super_over_match = icc_t20_df['Super Over Match'].value_counts()
super_over_match
```

```python
super_over_condition = icc_t20_df['Super Over Match'].str.contains('Yes', case=False, na=False)
no_result_condition = icc_t20_df['Match Result'].str.contains('No Result|Abandoned', case=False, na=False)
```

```python
super_over_count = icc_t20_df[super_over_condition].shape[0]
no_result_count = icc_t20_df[no_result_condition].shape[0]
normal_match_count = icc_t20_df[~super_over_condition & ~no_result_condition].shape[0]
```

```python
match_counts = pd.DataFrame({
    'Match Type': ['Super Over', 'Normal', 'No Result'],
    'Count': [super_over_count, normal_match_count, no_result_count]
})

plt.figure(figsize=(10, 6))
sns.barplot(x='Match Type', y='Count', data=match_counts, palette='hls')

plt.title('Counts of Super Over, Normal, and No Result Matches')
plt.xlabel('Match Type')
plt.ylabel('Count')
plt.ylim(0, max(match_counts['Count']) + 10)
```

```python
# Total Matches Played
matches_played = pd.concat([icc_t20_df['1st Team'], icc_t20_df['2nd Team']]).value_counts().reset_index()
matches_played.columns = ['Team', 'Matches Played']

plt.figure(figsize=(10,6))
sns.barplot(x='Matches Played', y='Team', data=matches_played, palette='viridis')
plt.title('Total Matches Played by Each Team')
plt.xlabel('Matches Played')
plt.ylabel('Team')
```

```python
# Create a win/loss count for each team
win_counts = icc_t20_df['Winners'].value_counts()
loss_counts = icc_t20_df['Losers'].value_counts()

team_results = pd.DataFrame({'Wins': win_counts, 'Losses': loss_counts}).fillna(0)

team_results['Win/Loss Ratio'] = team_results.apply(lambda x: x['Wins'] / x['Losses'] if x['Losses'] != 0 else x['Wins'], axis=1)

team_results = team_results.sort_values(by='Win/Loss Ratio', ascending=False).reset_index().rename(columns={'index': 'Team'})

plt.figure(figsize=(12, 6))
sns.barplot(x='Win/Loss Ratio', y='Team', data=team_results, palette='viridis')
plt.title('Win/Loss Ratio for Each Team')
plt.xlabel('Win/Loss Ratio')
plt.ylabel('Team')
```

```python
# top 4 teams with most wins
top_4_teams = team_results.nlargest(4, 'Wins')

plt.figure(figsize=(10,6))
sns.barplot(x='Wins', y='Team', data=top_4_teams, palette='viridis')
plt.title('Top 4 Teams with Most Wins')
plt.xlabel('Wins')
plt.ylabel('Team')
```

```python
# top 4 teams with most loss

filtered_df = icc_t20_df[icc_t20_df['Losers'] != 'No Result (Rain)']

loss_counts = filtered_df['Losers'].value_counts().nlargest(4).reset_index()
loss_counts.columns = ['Team', 'Losses']

plt.figure(figsize=(10,6))
sns.barplot(x='Losses', y='Team', data=loss_counts, palette='viridis')
plt.title('Top 4 Teams with Most Losses')
plt.xlabel('Losses')
plt.ylabel('Team')
```

```python
# Rename columns for clarity
runs_scored = icc_t20_df[['1st Team', 'First Innings Score']].rename(columns={'1st Team': 'Team', 'First Innings Score': 'Runs'})

runs_scored['Runs'] = pd.to_numeric(runs_scored['Runs'], errors='coerce')

total_runs = runs_scored.groupby('Team')['Runs'].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(x='Runs', y='Team', data=total_runs.sort_values(by='Runs', ascending=False), palette='inferno')
plt.title('Total Runs Scored by Each Team')
plt.xlabel('Total Runs')
plt.ylabel('Team')
```

```python
# Rename columns for clarity
runs_conceded = icc_t20_df[['2nd Team', 'First Innings Score']].rename(columns={'2nd Team': 'Team', 'First Innings Score': 'Runs'})

runs_conceded['Runs'] = pd.to_numeric(runs_conceded['Runs'], errors='coerce')

average_runs_conceded = runs_conceded.groupby('Team')['Runs'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(x='Runs', y='Team', data=average_runs_conceded.sort_values(by='Runs', ascending=False), palette='viridis')
plt.title('Average Runs Conceded by Each Team')
plt.xlabel('Average Runs Conceded')
plt.ylabel('Team')
```

```python
match_results_by_stage = icc_t20_df.melt(id_vars=['Stage'], value_vars=['Winners', 'Losers'], var_name='Result Type', value_name='Team')

plt.figure(figsize=(14, 8))
sns.countplot(y='Stage', hue='Result Type', data=match_results_by_stage, palette='viridis', order=icc_t20_df['Stage'].value_counts().index)
plt.title('Distribution of Match Results by Stage')
plt.xlabel('Number of Matches')
plt.ylabel('Stage')
plt.legend(title='Result Type')
```

```python
# Toss Analysis
toss_decisions = icc_t20_df['Toss Decision'].value_counts().reset_index()
toss_decisions.columns = ['Decision', 'Count']

plt.figure(figsize=(8, 6))
sns.barplot(x='Count', y='Decision', data=toss_decisions, palette='viridis')
plt.title('Toss Decisions')
plt.xlabel('Count')
plt.ylabel('Decision')
```

```python
def plot_toss_impact_on_match_outcome(df):
    df['Toss Impact'] = df['Toss Winning'] == df['Winners']
    toss_impact_counts = df['Toss Impact'].value_counts().reset_index()
    toss_impact_counts.columns = ['Toss Impact', 'Count']
    
    plt.figure(figsize=(8, 6))
    sns.barplot(x='Toss Impact', y='Count', data=toss_impact_counts, palette='viridis')
    plt.title('Impact of Winning Toss on Match Outcome')
    plt.xlabel('Toss Winner Same as Match Winner')
    plt.ylabel('Number of Matches')
    plt.xticks([0, 1], ['No', 'Yes'])
    plt.show()

plot_toss_impact_on_match_outcome(icc_t20_df)
```
